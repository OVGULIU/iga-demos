### function dimsplinespace(p::Integer, kts::Vector{Float64})

```julia
n = dimsplinespace(p,kts)
println("The dimension of the spline space is: n = $n")
```

### function grevillepoints(p::Integer, kts::Array{Float64,1})

```julia
# compute the Greville Absiscae of the spline space defined by degree 'p' and knotvector 'kts'
#
# input:
#    p   :: Int             - polynomial degree
#    kts :: Vector{Float64} - knotvector
# output:
#    y   :: Vector{Float64} - vector with Greville Absiscae
```

```julia
y = grevillepoints(p,kts)
```

### function onebasisfuneval(p::Integer, kts::Vector{Float64}, u::Float64)

```julia
# example input and output
p = 2
knots = [0.0,1.0,2.0,3.0]
u = 1.5
b = onebasisfuneval(p,knots,u)
println("The value of the basis function evaluated at u = $u is given by: b = $b")
```

### function onebasisfuneval(p::Integer,kts::Vector{Float64},u::Vector{Float64})

```julia
# compute value of a single B-spline basis-function of degree 'p' with local knotvector 'knots' evaluated at points 'u'
#
# input:
#    p     :: Int             - polynomial degree
#    knots :: Vector{Float64} - knotvector
#    u     :: Vector{Float64} - vector with evaluation points
#
# Output: 
#    b     :: Vector{Float64} - basisfunction defined by degree 'p' and local knotvector 'knots' evaluated at points 'u'
```

```julia
# example input and output
p = 2
knots = [0.0,1.0,2.0,3.0]
u = linspace(knots[1],knots[end],100)
println("The value of the basis function evaluated at the points u is given by:")
b = onebasisfuneval(p,knots,u)
```

```julia
# example use
plot(u,b)
```

### function bsplinebasiseval(p::Integer, kts::Vector{Float64}, span::Integer, u::Float64)

```julia
# example of inpurt and output
u    = 0.45                               # evaluation point
span = findspan(p,kts,u)                  # span-index of point u in knotvector 
Nu, supp = bsplinebasiseval(p,kts,span,u) # compute B-spline basis and corresponding support
```

```julia
# example of use
B = Nu[:,p+1]; Bsupp = supp[:,p+1]
println("The pth degree B-spline basis evaluated at u = $u is: $B")
println("Its corresponding support is: $Bsupp")
```

### function bsplinebasiseval(p::Integer, kts::Vector{Float64}, u::Float64)

```julia
# similar use as the above function - no need to specify the knotspan index
u = 0.45
Nu, supp = bsplinebasiseval(p,kts,u)
```

### function dersbsplinebasisfuns(p::Degree, kts::KnotVector, u::Float64, span::Int, nout::Int)

```julia
# example input and output
u    = 0.45                               # evaluation point
span = findspan(p,kts,u)                  # span-index of point u in knotvector 
B    = dersbsplinebasisfuns(p, kts, span, u, 3)
```

```julia
# example of use
b = B[:,1]; ∇b = B[:,2]; Δb = B[:,3];
println("The pth degree B-spline basis evaluated at u = $u is: $b")
println("Its 1st derivatives evaluated at u = $u are: $∇b")
println("Its 2nd derivatives evaluated at u = $u are: $Δb")
```

### function dersbsplinebasisfuns(p::Degree, kts::KnotVector, u::Float64, nout::Int)

```julia
# similar use as the above funtion - no need to specify knot-span index
u    = 0.45
b    = dersbsplinebasisfuns(p, kts, u, 3)
```

### function dersbsplinebasisfuns(p::Int, kts::Vector{Float64}, span::Vector{Int64}, u::Vector{Float64}, nout::Int)

```julia
# example input and output
u    = grevillepoints(p,kts)                   # vector with evaluation points
span = findspan(p,kts,u)                       # span-index of point u in knotvector 
B    = dersbsplinebasisfuns(p, kts, span, u, 3)
```

```julia
# example of use
println("The (p+1 × length(u)) matrix of B-spline basis functions is:")
B[1]
```

```julia
# example of use
println("The (p+1 × length(u)) matrix of B-spline 1st derivative functions is:")
B[2]
```

```julia
# example of use
println("The (p+1 × length(u)) matrix of B-spline 2nd derivative functions is:")
B[3]
```

### function dersbsplinebasisfuns(p::Int, kts::Vector{Float64}, u::Vector{Float64}, nout::Int)

```julia
# similar use as above function - no need to specify knot span index
# example input and output
u    = grevillepoints(p,kts)                   # vector with evaluation points
B    = dersbsplinebasisfuns(p, kts, u, 3)
```

### function dersbsplinebasisfunsinterpolationmatrix(p::Int, kts::Vector{Float64}, u::Vector{Float64}, nout::Int)

```julia
# compute B-spline sparse collocation matrices of the basis of degree 'p' and its '1,2,...,nout-1'th order derivatives at the vector of sites 
#
# input:
#    p     :: Int             - polynomial degree
#    kts   :: Vector{Float64} - knotvector
#    u     :: Vector{Float64} - vector with evaluation points
#    nout  :: Int             - number of outputs
# Output: 
#    ders  :: Vector{Matrix{Float64}} - [B(u); B'(u); B''(u) ... ] where B(u) denotes the collocation matrix of the B-spline basis functions 
#                                       evaluated at the vector of sites u
```

```julia
# similar use as above function - no need to specify knot span index
# example input and output
u    = linspace(kts[1],kts[end],1000)
B    = dersbsplinebasisfunsinterpolationmatrix(p, kts, u, 3)
```

```julia
figure(); plot(u,B[1]);
figure(); plot(u,B[2]);
figure(); plot(u,B[3]);
```
